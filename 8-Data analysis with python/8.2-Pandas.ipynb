```python
## Let's discuss about Series
## Series --> This is like a 1d array in numpy, that can hold any datatype. similar to column in a table
##            This is a labelled array, where as list is an unlabelled array.

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)

print(series)

data2 = [6,7,8,9]
ser = pd.Series(data2, index = ['a','b','c','d'])

print(ser)
```

```python
## Dataframe --> Similar to the series but this is a 2 dimensional, mutable, and heterogenous tabular data
# Creating a df using a dictonary

data = {
    'Name': ['Aladdin','Alibaba'],
    'Post':['Magi','King']
}

df = pd.DataFrame(data)
print(df)
```
