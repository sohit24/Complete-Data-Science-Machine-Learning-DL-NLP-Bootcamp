Seaborn is a python library based on matplotlib which is a high level interface for drawing attractive and informative graphics

```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
## for this I am going to use car_crashes dataset
crash_df = sns.load_dataset('car_crashes')
crash_df
```

Distribution Plot

```python
sns.displot(crash_df['not_distracted'], bins = 20) ### bins is not necessary it just divides the data between 20 bars
```

Joint Plot

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'hex')
```

Pair plot


```python
sns.pairplot(crash_df)
```

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
```

```python
tips_df
```

```python
sns.rugplot(tips_df['tip'])
```

```python
import matplotlib.pyplot as plt

sns.jointplot(x='speeding', y= 'alcohol', data=crash_df, kind='reg')

sns.set_style('whitegrid')
sns.set_context('paper', font_scale=1.4) ### there are different types of contexts where every context makes the graph different and readable
plt.figure(figsize=(10,5)) ### size of the chart

### We can also remove the axes in the graph using the despine() function and sns
```

Categorical Plots

```python
### Bar Plot
# sns.barplot(x='sex', y='tip',data=tips_df)
sns.barplot(x='sex', y='smoker', data=tips_df)
```

```python
### Count plot
sns.countplot(x='smoker', data=tips_df)
```

```python
### Box Plot
### plot tip by day on basis of gender
sns.boxplot(data=tips_df, x='tip', y='day', hue='sex')
```

```python
### Violin Plot
### plot between total_bill and the day on basis of gender
sns.violinplot(data=tips_df, x='day', y='total_bill', hue='sex',split=True) ## Split = True isn't necessary if you want them seperate don't include them
```
